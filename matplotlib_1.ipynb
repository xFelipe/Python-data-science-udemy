```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x * x
y
```

```python
# funcional
plt.plot(x, y,'r--') #red tracejado
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
##plt.show()
```

```python
#plt.subplots(nrows=1, ncols=2)
plt.subplot(1, 2, 1) #Uma linha, 2 colunas, usando elemento 1
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2) #Uma linha, 2 colunas, usando elemento 2
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.35, 0.3]) #distancia pra esquerda, distancia de baixo, largura, altura

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y, x)
```

```python
fig, ax = plt.subplots() # Cria a figura e o eixo para comportar dados
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax  # Lista de eixos
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Título')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Título2')

ax[1].plot(x, y, 'b')
ax[1].set_title('Título2')
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()  # Organiza e cuida das sobreposições dos gráficos
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,5))

axes.plot(x, y, 'r--', label='dados 1')
axes.plot(y, x, 'g--', label='dados 2')
axes.set_title('Título')
axes.legend(loc=0)  # Mostra legenda(Label)
```
